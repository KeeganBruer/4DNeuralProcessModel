```python
import plotly.express as px
import plotly.graph_objects as go
import math
import numpy as np
```

```python
training_data_loc = "./training_data/"
data = np.load(training_data_loc + "sample_set_0.npz")
sample_set_x = data["sample_set_x"]
points = []
for i in range(3585+285):#len(sample_set_x)):
    point_set = sample_set_x[i]
    point1 = point_set[:3]
    point2 = point_set[4:7]
    points.append(point1)
    points.append(point2)
points = np.array(points)
```

```python
fig = go.Figure()
#fig.update_xaxes(range=[140, 190])
#fig.update_traces(mode='markers')
fig.add_trace(go.Scatter3d(
    x=points[:,0],
    y=points[:,1],
    z=points[:,2],
    mode='markers',
))

fig.show()
```
